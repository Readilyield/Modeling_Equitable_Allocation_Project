```python
# =================================
# POP_DIR = cur_path+"/Dataset/population_data"
# CSV_IN  = cur_path+"/Dataset/data_complete.csv"
# CSV_OUT = cur_path+"/Dataset/data_with_age_groups.csv"
# =================================


def extract_age_groups(xlsx_path):
    """
    For a given state Excel file (like 'Kansas.xlsx'):
    - locate the column where Year == 2020 and Sex == 'Both Sexes'
    - age_1 = sum of rows 11–19 (Excel rows) in that column
    - age_2 = sum of rows 20–24 in that column

    Returns (age_1, age_2) as numbers.
    """
    # Read without headers so we can treat all rows uniformly
    df = pd.read_excel(xlsx_path, sheet_name=0, header=None)

    # 1) Find the row containing the '2020' year header
    #    One of the cells in that row will be 2020
    year_row_idx = df.isin([2020]).any(axis=1).idxmax()
    sex_row_idx = year_row_idx + 1

    year_row = df.iloc[year_row_idx]
    sex_row = df.iloc[sex_row_idx]

    # 2) Find the column where year == 2020 and sex == "Both Sexes"
    mask = (year_row == 2020) & (sex_row == "Both Sexes")
    cols = df.columns[mask]
    if len(cols) == 0:
        raise ValueError(f"Could not find 2020 / Both Sexes column in {xlsx_path}")
    col = cols[0]

    # 3) Sum the specified rows in that column.
    #    Excel rows 11–19 -> pandas indices 10–18 (inclusive)
    #    Excel rows 20–24 -> pandas indices 19–23 (inclusive)
    age_1 = df.iloc[10:19, col].sum()  # rows 11–19
    age_2 = df.iloc[19:24, col].sum()  # rows 20–24

    return age_1, age_2


# Build mappings: State -> age_1, State -> age_2
# age_1_map = {}
# age_2_map = {}

# for path in glob.glob(os.path.join(POP_DIR, "*.xlsx")):
#     filename = os.path.basename(path)
#     # Example: 'Kansas.xlsx' -> 'Kansas'
#     #          'New Hampshire.xlsx' -> 'New Hampshire'
#     state_name = os.path.splitext(filename)[0].replace("_", " ")

#     try:
#         a1, a2 = extract_age_groups(path)
#         age_1_map[state_name] = a1
#         age_2_map[state_name] = a2
#         print(f"{state_name}: age_1={a1}, age_2={a2}")
#     except Exception as e:
#         print(f"Skipping {filename} due to error: {e}")


# Load the existing dataset
# df = pd.read_csv(CSV_IN)

# # Add the new columns by mapping from the State column
# df["age_1"] = df["State"].map(age_1_map)
# df["age_2"] = df["State"].map(age_2_map)

# # Optional: check which states didn’t get matched
# missing = df[df["age_1"].isna() | df["age_2"].isna()]["State"].unique()
# if len(missing) > 0:
#     print("States with missing age_1 / age_2 (no matching Excel or parse issue):")
#     print(missing)

# # Save the updated CSV
# df.to_csv(CSV_OUT, index=False)
# print(f"Saved dataset with age_1 and age_2 to: {CSV_OUT}")
```

### Plot US map and locations

```python
states_in_data = df["State"]
shp_path = "Dataset/map/cb_2024_us_state_20m.shp"

gdf_states = gpd.read_file(shp_path)

# Make sure we're starting in WGS84 (lon/lat)
if gdf_states.crs is None:
    gdf_states = gdf_states.set_crs(epsg=4326)
else:
    gdf_states = gdf_states.to_crs(epsg=4326)

all_states = gdf_states[~gdf_states["NAME"].isin(["Alaska", "Puerto Rico"])]
# Project all_states to an equal-area US CRS
all_states_proj = all_states.to_crs(epsg=2163)  # US National Atlas Equal Area

# Centroids in projected CRS
all_states_proj["centroid"] = all_states_proj.geometry.centroid

# After computing centroids for all_states:
centroids = (
    all_states_proj[["NAME", "centroid"]]
    .set_geometry("centroid")
    .to_crs(epsg=4326)
)

centroids["cen_lon"] = centroids.geometry.x
centroids["cen_lat"] = centroids.geometry.y

tgt_centroids = centroids[centroids["NAME"].isin(states_in_data)]

# origin place coordinates
kalamazoo_lat = 42.291707
kalamazoo_lon = -85.587229

# plot
fig, ax = plt.subplots(figsize=(12, 10))

# Plot the US state outlines (shapefile-based)
all_states.boundary.plot(ax=ax, linewidth=0.5, linestyle="--",color="black")
tgt_states.boundary.plot(ax=ax, linewidth=1.0)

# Plot Kalamazoo
ax.scatter(kalamazoo_lon, kalamazoo_lat, s=0, 
           label="This is not a complete map")
ax.scatter(kalamazoo_lon, kalamazoo_lat, s=25, color="green")
ax.text(kalamazoo_lon, kalamazoo_lat + 0.5, "Origin", ha="center", color="green")

# Draw arrows from Kalamazoo to each state centroid
for _, row in tgt_centroids.iterrows():
    lon = row["cen_lon"]
    lat = row["cen_lat"]
    state_name = row["NAME"]

    # Arrow from Kalamazoo to the state centroid
    ax.annotate(
        "",
        xy=(lon, lat),
        xytext=(kalamazoo_lon, kalamazoo_lat),
        arrowprops=dict(arrowstyle="-", linewidth=0.5, alpha=0.6),
    )

    # Mark centroid + label
    ax.scatter(lon, lat, color="orange",s=12)
#     ax.text(lon, lat - 0.7, state_name, fontsize=6, ha="center")

# Tidy up axes: focus on US
ax.axis("off")
ax.set_title("From Origin to Locations in Need",fontsize=14)

ax.set_aspect(1)  # make degrees look roughly square
ax.grid(True, linewidth=0.3, alpha=0.5)
ax.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
